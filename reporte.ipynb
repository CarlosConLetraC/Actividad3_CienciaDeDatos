## 📊 Métricas del modelo

```python
import json

with open('data/actividad3.json') as f:
    data = json.load(f)

m = data['metrics']

print('Pearson:', m['pearson'])
print('RMSE:', m['rmse'])
print('Error medio:', m['error_media'])
print('Desviación estándar:', m['error_std'])
print('Varianza:', m['error_var'])
```

## 📌 Conclusión

El modelo de regresión lineal simple muestra una correlación positiva moderada entre hits y runs.

- Pearson indica relación lineal existente
- RMSE refleja error moderado esperado en datos deportivos

📌 Conclusión técnica:
El modelo es útil para estimaciones generales, pero no para predicciones exactas debido a la variabilidad inherente del deporte.
